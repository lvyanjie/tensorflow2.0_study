```python
import cv2
import matplotlib.pyplot as plt
# 验证kernel
img = cv2.imread('20190126105450630.jpeg')
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
import numpy as np
kernel = np.array([[0, 0, 0],
                  [0, 1, 0],
                  [0, 0, 0]])

img_1 = cv2.filter2D(img, -1, kernel)
plt.imshow(img_1[:,:,::-1])
plt.show()
```

```python
kernel = np.array([[0, -1, 0],
                  [-1, 5, -1],
                  [0, -1, 0]])

img_2 = cv2.filter2D(img, -1, kernel)
plt.imshow(img_2[:,:,::-1])
plt.show()
```

```python
kernel = np.array([[0.625, 0.125, 0.625],
                  [0.125, 0.25, 0.125],
                  [0.625, 0.125, 0.625]])

img_3 = cv2.filter2D(img, -1, kernel)
plt.imshow(img_3[:,:,::-1])
plt.show()
```

```python
kernel = np.array([[-1, -1, -1],
                  [-1, 8, -1],
                  [-1, -1, -1]])

img_4 = cv2.filter2D(img, -1, kernel)
plt.imshow(img_4[:,:,::-1])
plt.show()
```
